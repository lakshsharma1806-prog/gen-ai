```python
#non keywoed variable length argument
def test(*args):
  print(args)
  print(len(args))
  print(type(args))
```

```python
test(2, 3, 4, 5, 6)
```

```python
def sum_number(*args):
  sum = 0
  for num in args:
    sum+=num
  return sum
```

```python
sum_number(1, 2, 3, 4, 5, 6)
```

```python
#keyword variable length argument()
def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
test(a = 10, b = 20, c = 30 )
```

```python
def pass_student(**kwargs):
  pass_student_list = []
  for name, marks in kwargs.items():
    if marks>33:
      print(name, marks)
  return pass_student_list
```

```python
pass_student(A=10, B=20, C=30, D=40, E=50, F=60, G=70, H=80)
```

```python
cgpa = np.array([6,6.5,7,8,9,9.9,10.0,6.7])
package = np.array([4,4.5,6,6.2,8,4,5.5, 5.0])
font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'serif', 'color':'darkred', 'size':20}
plt.plot(cgpa, package, "o:y", ms = 10, mec = "Orange", mfc = "Red")
plt.xlabel("cgpa")
plt.ylabel("package")
plt.title("CGPA VS PACKAGE")
plt.show()
```

```python
#bar plot-- it work on numerical as well as categorial data
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([800, 2000, 3000, 2309])
color = np.array(["Red", "Yellow", "Orange", "Green"])
plt.bar(fruit, quantity, color = color)
plt.show()
```
